```python
import numpy as np
import itertools
from collections import deque

cards = np.arange(1, 8)
investors = np.arange(1, 8)

def num_match(cards, investors):
    return sum(np.equal(cards, investors))

def highest_num_match(cards, investors):
    rotated_investors = deque(investors)
    h_num_match = 0
    for i in range(7):
        rotated_investors.rotate(-1)
        nm = num_match(cards, rotated_investors)
        if(nm > h_num_match):
            h_num_match = nm
    return h_num_match

def try_all_permuations(cards, investors):
    for perm_investors in list(itertools.permutations(investors)):
        n = highest_num_match(cards, np.array(perm_investors))
        if n <= 1:
            print(f"Permutation: {perm_investors}")
        

try_all_permuations(cards, investors)
```
